```python
import pandas as pd 
from pathlib import Path
import hvplot.pandas
%matplotlib inline
```

```python
 #Set file paths
file_path = Path("../data/Total_Energy_Use.csv")

# Read CSV as a DataFrame
consumption_data = pd.read_csv(file_path)

#consumption_data.set_index(consumption_data['Year'], inplace=True)

consumption_data = consumption_data


consumption_data.head()
```

```python
# Grab the 'Crude Oil and Lease Condensate quads', 'Natural Gas Plant Liquids quads', 'Coal quads', 'Nuclear quads', 'Hydropower quads,'Other Renewable quads', columns

consumption = consumption_data.loc[:,['Year','Liquid Fuels quads', 'Natural Gas quads','Coal quads', 'Nuclear quads', 'Hydropower quads', 'Other Renewable Energy quads']]

#consumption.sort_index(ascending=True)

consumption.head()

```

```python
# Data Prep and plotting

consumption.hvplot.line(
    title= "Total Energy Consumption", 
    x='Year', 
    y=[ 'Liquid Fuels quads', 'Natural Gas quads','Coal quads', 'Nuclear quads','Hydropower quads','Other Renewable Energy quads']

)
```

```python
# Subplotting each energy consumption

consumption.hvplot(
    x='Year', 
    y=['Liquid Fuels quads', 'Natural Gas quads','Coal quads', 'Nuclear quads','Hydropower quads','Other Renewable Energy quads'], 
    subplots=True, width=500, height=300, rot=90,

)

#will not use for presentation
```
