```python
# Import libraries and dependencies
import pandas as pd
from pathlib import Path
import seaborn as sns
%matplotlib inline
```

Read CSV in as DataFrame

```python
#Energy Production data read in and "year" set as index
Energy_Production_data = Path("../data/Total_Energy_Production.csv") 
Energy_Production = pd.read_csv(Energy_Production_data, index_col="Year")
Energy_Production.head()
```

```python
# Non-renewable Energy Production
nonrenewable_production = Energy_Production.loc[:, ['Crude Oil and Lease Condensate quads', 'Natural Gas Plant Liquids quads', 'Dry Natural Gas quads', 'Coal quads','Nuclear quads', 'Other quads']]
nonrenewable_production.head()
```

```python
#Read in Energy consumption data and "year" set as index  
Energy_Consumption_data = Path("../data/Total_Energy_Use.csv")
Energy_Consumption = pd.read_csv(Energy_Consumption_data, index_col="Year")
Energy_Consumption.head()
```

```python
Nonrenewable_Consumption = Energy_Consumption.loc[:, ['Liquid Fuels quads', 'Natural Gas quads', 'Coal quads','Nuclear quads']]
Nonrenewable_Consumption.head()
```

```python
#Read in norminal price  set data and "year" as index  
Energy_Norminal_Price_data = Path("../data/Total_Energy_Nominal_Prices.csv")
Energy_Norminal_Price = pd.read_csv(Energy_Norminal_Price_data, index_col="Year")
Energy_Norminal_Price.head()
```

```python
Energy_Price = Energy_Norminal_Price.loc[:, ['West Texas Intermediate (nom $/b) nom $/b', 'Gas Price at Henry Hub (nom $/MMBtu) nom $/MMBtu', 'Coal at Minemouth (nom $/ton) nom $/ton', 'Electricity (nom cents/kWh) nom cents/kWh']]
Energy_Price.head()
```

```python
#Used concat to combine the Production and Norminal price data and idexed it by years
Prod_Price_combined_df = pd.concat([nonrenewable_production, Energy_Price], axis="columns", join="inner")
Prod_Price_combined_df
```

```python
#Used concat to combine the consumption and Norminal price data and idexed it by years
Consu_Price_combined_df = pd.concat([Nonrenewable_Consumption, Energy_Price], axis="columns", join="inner")
Consu_Price_combined_df
```

```python
#Used concat to combine the consumption and Production data and idexed it by years
Consu_Prod_combined_df = pd.concat([nonrenewable_production, Nonrenewable_Consumption], axis="columns", join="inner")
Consu_Prod_combined_df
```
