```python
import pandas as pd 
import plotly.express as px
from pathlib import Path
import hvplot.pandas
%matplotlib inline
```

```python
# Set file paths
file_path = Path("../data/Total_Energy_Production.csv")

# Read CSV as a DataFrame
production_data = pd.read_csv(file_path)
#production_data.set_index(production_data['Year'], inplace=True)

production_data = production_data.iloc[15:-1]

production_data.head(10)
   




```

```python
# Grab the 'Crude Oil and Lease Condensate quads', 'Natural Gas Plant Liquids quads', 'Coal quads', 'Nuclear quads', 'Hydropower quads,'Other Renewable quads', columns
Production = production_data.loc[:, [ 'Year','Crude Oil and Lease Condensate quads', 'Natural Gas Plant Liquids quads', 'Coal quads', 'Nuclear quads', 'Hydropower quads', 'Other Renewable Energy quads']]

Production.sort_index(ascending=True)

Production.head(15)
```

```python
# Data Prep plot
Production.hvplot.line(x='Year', y=['Crude Oil and Lease Condensate quads', 'Natural Gas Plant Liquids quads', 'Coal quads', 'Nuclear quads','Hydropower quads','Other Renewable Energy quads'])
```

```python
# suplots to show each energy

Production.hvplot(x='Year', y=['Crude Oil and Lease Condensate quads', 'Natural Gas Plant Liquids quads', 'Coal quads', 'Nuclear quads',	'Hydropower quads','Other Renewable Energy quads'],  subplots=True, 
width=400, height=300, rot=90)
```

```python
# Bargraph 

Production.hvplot.bar(x='Year', y=['Crude Oil and Lease Condensate quads', 'Natural Gas Plant Liquids quads', 'Coal quads', 'Nuclear quads','Hydropower quads','Other Renewable Energy quads'], value_labels ='Quads', height=600, width=1200, rot=90)
```
